```python
import torch
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#true model parameters
w1_true = 1.0
w2_true = 2.2
b_true = -0.5

#generate synthetic data
npts = 200
#x = np.random.normal(loc=0.0,scale=4.0,size=npts)
x = np.random.random(npts)*10 - 5
x.sort() #not necessary, aranges the x values from low to high
noise = np.random.normal(loc=0.0,scale=3.0,size=npts)
y = w1_true*x + w2_true*x**2 + b_true + noise

#initialize weight(s) and bias(es) as random number
w1 = np.random.random()
w2 = np.random.random()
b = np.random.random()

#hyper parameters
learning_rate = 1e-5 #learning rate
nepochs = 500

#main training loop
all_loss = []
for i in range(0,nepochs):

    y_pred = w1*x + w2*x**2 + b
    loss = 0.5*np.square(y_pred - y).sum()
    grad_y_pred = (y_pred - y)
    grad_b = grad_y_pred.sum()
    grad_w1 = (grad_y_pred*x).sum()
    grad_w2 = (grad_y_pred*x**2).sum()
    all_loss.append(loss)

    b -= learning_rate * grad_b
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2

    if i % 100 == 0:
        print('epoch: {}, loss: {}'.format(i,loss))
        
print('\n')
print('w1 = {}, w1_true = {}'.format(w1,w1_true))
print('w2 = {}, w1_true = {}'.format(w2,w2_true))
print('b = {}, b_true = {}'.format(b,b_true))
```
